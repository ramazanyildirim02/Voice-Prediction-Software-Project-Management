# mp3 uzantılı ses dosyasını wav uzantılı ses dosyasına çevirme

```python
from pydub import AudioSegment

def convert_m4a_to_wav(input_path, output_path):
    # M4A dosyasını yükle
    audio = AudioSegment.from_file(input_path, format="m4a")

    # WAV formatına dönüştür ve kaydet
    audio.export(output_path, format="wav")

# M4A dosyasının bulunduğu dizin
m4a_file_path = "/Users/ramazanyildirim/Desktop/ses_projesi/ml_model/ses_dosyaları/caltepe.m4a"

# WAV olarak kaydedilecek dosyanın adı ve yolu
wav_file_path = "/Users/ramazanyildirim/Desktop/ses_projesi/ml_model/ses_dosyaları/caltepe.wav"

# Dönüştürme işlemini gerçekleştir
convert_m4a_to_wav(m4a_file_path, wav_file_path)

print(f"{m4a_file_path} dosyası {wav_file_path} dosyasına dönüştürüldü.")
```

```python
import librosa
import librosa.display
import IPython.display as ipd
import pandas as pd
import numpy as np

def extract_features(file_path):
    # Ses dosyasını yükle
    y, sr = librosa.load(file_path)
        
    # Mel-frekans spektrogramunu hesapla
    mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)

    # Vektörleştirme (örneğin, sütunlardaki ortalamayı alabilirsiniz)
    feature_vector = np.mean(mel_spectrogram, axis=1)

    return feature_vector

# Ses dosyalarının bulunduğu dizini belirtin
ses_dosyalari_dizini = "/Users/ramazanyildirim/Desktop/ses_projesi/ml_model/ses_dosyaları"

# Ses dosyalarını işleyip dataframe'e ekleyin
#es_dosyasi_listesi = ["ceylin1.wav", "ceylin2.wav","ceylin3.wav","ceylin4.wav","ceylin5.wav","ceylin6.wav","ceylin7.wav","ceylin8.wav","ceylin9.wav","ceylin10.wav","ceylin11.wav","ceylin12.wav","ceylin13.wav"]  # Ses dosyalarının adlarını içeren bir liste
#ses_dosyasi_listesi = [ "semih1.wav","semih2.wav","semih3.wav","semih4.wav","semih5.wav","semih6.wav","semih7.wav","semih8.wav","semih9.wav","semih10.wav","semih11.wav"]
#ses_dosyasi_listesi = [ "hasan1.wav", "hasan2.wav", "hasan3.wav", "hasan4.wav", "hasan5.wav", "hasan6.wav", "hasan7.wav", "hasan8.wav", "hasan9.wav", "hasan10.wav", "hasan11.wav"]
ses_dosyasi_listesi = [ "ramazan1.wav", "ramazan2.wav", "ramazan3.wav","ramazan4.wav", "ramazan5.wav","ramazan6.wav","ramazan7.wav","ramazan8.wav", "ramazan9.wav","ramazan10.wav","ramazan11.wav"]      

data = []
for ses_dosyasi in ses_dosyasi_listesi:
    dosya_yolu = ses_dosyalari_dizini + "/" + ses_dosyasi
    features = extract_features(dosya_yolu)
    data.append(features)

# DataFrame oluştur
columns = [f"feature_{i+1}" for i in range(len(data[0]))]
df = pd.DataFrame(data, columns=columns)

# Ses dosyası adlarını dataframe'e ekleyebilirsiniz
#df["ses_dosyasi"] = ses_dosyasi_listesi

df["ses_dosyasi"] = "ramazan"

```

```python
import matplotlib.pyplot as plt
def load_hist(x,yol,isim):
    x, sr = librosa.load(yol)
    
    plt.figure(figsize=(15,17))
    plt.subplot(3,1,1)
    plt.plot(x)
    plt.title(f"{isim} ait tahmin")
    plt.ylim((-1,1))
    plt.show()
    
```
